##**Parkinson**

El Parkinson es una enfermedad neurodegenerativa crónica que afecta el sistema nervioso central y se caracteriza principalmente por la degeneración de células nerviosas que producen dopamina en el cerebro. Esta condición provoca síntomas motores como temblores, rigidez muscular y dificultad para el movimiento. Además de los síntomas motores, el Parkinson puede manifestar efectos no motores, como trastornos cognitivos y emocionales.

En la investigación del Parkinson, uno de los enfoques prominentes implica el uso de la electroencefalografía (EEG). Esta técnica registra la actividad eléctrica del cerebro mediante electrodos colocados en el cuero cabelludo. El EEG se ha empleado en estudios para analizar las señales cerebrales en pacientes con Parkinson, con el propósito de profundizar en la comprensión de la enfermedad y avanzar en el desarrollo de tratamientos más efectivos.

# 1. Implemente una función que reciba una señal de múltiples canales y épocas y calcule la Energía de promedio de cada canal.

```python
import numpy as np

def calcular_energia_promedio(señal):
    # Calcular la energía de cada época y canal
    energias_por_epoca = np.sum(np.square(señal), axis=1)

    # Calcular la energía promedio de cada canal
    energia_promedio_por_canal = np.mean(energias_por_epoca, axis=0)

    return energia_promedio_por_canal

# Ejemplo de uso:


print("Energía promedio por canal para el grupo de pacientes con Parkinson:")
print(calcular_energia_promedio(senales_control[0][0,:,:]))
print("Energía promedio por canal para el grupo control:")
print(calcular_energia_promedio(senales_parkinson[0][0,:,:]))
```

# 3. Determine si existe diferencia estadística entre canales de cada grupo de sujetos a través de una prueba t. Compruebe los supuestos necesarios para realizar una prueba t, esto es: Normalidad de la variable, independencia (se asume que los grupos son independientes), y homocedasticidad (use una prueba de Levene), finalmente realice la prueba t para determinar si existen diferencias entre los canales entre grupos de sujetos. De no cumplirse los requisitos, realice entonces un análisis no paramétrico (prueba U de Mann-Whitney). Este numeral tiene como objetivo identificar los canales que entregan información diferencial entre pacientes Sanos y con enfermedad de Parkinson.

```python
def normalidad(señal):
  shapiro=stats.shapiro(señal)
  pvalue=shapiro.pvalue
  #h1=las muestras no distribuyen normal
  #h0= las muestras distribuyen normal
  # si el valor p es menor a 0.05 se rechaza la hipotesis nula
  if pvalue <= 0.05:
    resultado = "los datos no distribuyen normalmente (p = {:.50f})".format(pvalue)
  else:
      resultado = "los datos distribuyen normalmente (p = {:.4f})".format(pvalue)
  return resultado,shapiro

def homocedasticidad(signal1,signal2):
    homo=stats.levene(signal1,signal2)
    pvalue=homo.pvalue
  # h0='las varianzas son iguales'
  # h1= 'las varianzas son diferentes'
  #cuando p<= 0.05 se rechaza h0, de lo contrario se acepta
    if pvalue < 0.05:
      resultado = "las  varianzas son diferentes (p = {:.50f})".format(pvalue)
    else:
        resultado = "las varianzas son iguales (p = {:.4f})".format(pvalue)

    return resultado

def pruebaT(s1,s2):
  test=stats.ttest_ind(s1,s2)
  #h0= no hay diferencias entre las dos medias
  #h1= hay diferencias significativas entre las dos medias
  #cuando p<= 0.05 se rechaza h0, de lo contrario se acepta

  pvalue=test.pvalue
  if pvalue <= 0.05:
    resultado = "Hay diferencias en las medias (p = {:.50f})".format(pvalue)
  else:
      resultado = "No hay diferencias en las medias (p = {:.4f})".format(pvalue)

  return resultado

def mann_w(signal1,signal2):
  mann=stats.mannwhitneyu(signal1,signal2)
  pvalue=mann.pvalue
  #h0=no hay diferencias entre los dos ciclos extraídos de la señal filtrada.
  #h1= existe diferencias entre los dos ciclos extraídos de la señal filtrada.
  if pvalue <= 0.05:
    resultado = "Hay diferencias entre los dos ciclos control y parkinson (p = {:.50f})".format(pvalue)
  else:
      resultado = "No hay diferencias entre los dos ciclos control y parkinson (p = {:.4f})".format(pvalue)

  return resultado
```

# PRUEBA DE NORMALIDAD

```python
labels1=['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8']
for i in labels1:
  c=control[f'{i}']
  cp=parkinson[f'{i}']
  a=normalidad(c)[0]
  b=normalidad(cp)[0]
  print(f"En el canal {i} control " + a + f"\nEn el canal {i} parkinson " + b+'\n')
```

```python
for i in labels1:
  c=control[f'{i}']
  cp=parkinson[f'{i}']
  a=homocedasticidad(c,cp)
  print(f'En los ciclos {i} de control y parkinson '+a)
```

Basándonos en los valores p obtenidos, podemos afirmar que los datos no siguen una distribución normal para cada canal y grupo (control y Parkinson). Los valores p son significativamente bajos (menores a 0.05), lo que indica que hay una diferencia significativa entre la distribución de los datos y una distribución normal. Por lo tanto no cumplimos con el supuesto de normalidad necesario para realizar una prueba paramétrica, como la prueba t.


Al no cumplirse los requisitos para realizar la prueba t, realizamos entonces un análisis no paramétrico (prueba Ude Mann-Whitney).

```python
# Crear una lista para almacenar los resultados de la prueba de Mann-Whitney U para cada canal
resultados_mannwhitney = []

# Iterar sobre cada canal en labels1
for i in labels1:
    # Obtener los datos del canal i para control y parkinson
    c = control[i]
    cp = parkinson[i]

    # Realizar la prueba de Mann-Whitney U para los datos del canal i
    resultado_mannwhitney = mann_w(c, cp)

    # Agregar el resultado de la prueba a la lista de resultados
    resultados_mannwhitney.append((i, resultado_mannwhitney))

# Imprimir los resultados
for canal, resultado in resultados_mannwhitney:
    print(f"Canal {canal}: {resultado}")
```

Los resultados indican que no se encontraron diferencias significativas entre pacientes sanos y con enfermedad de Parkinson en ninguno de los canales analizados. Todos los valores p son mayores que 0.05, lo que sugiere que no hay evidencia suficiente para rechazar la hipótesis nula de que no hay diferencias entre los dos grupos en cada canal.

Esto podría implicar que los canales analizados no son capaces de proporcionar información diferencial entre los dos grupos de pacientes en este estudio específico. Es posible que se necesiten otros métodos de análisis o que se deban considerar diferentes características de las señales para detectar diferencias entre los grupos.
