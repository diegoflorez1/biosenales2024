#PUNTO 1

````python
# Definición de la función que genera la función escalón unitario en un intervalo dado
def escalon_unitario(n0, n1, n2):
    # Inicializar una lista vacía para almacenar los valores del escalón
    escalon = []
    # Iterar sobre cada valor en el rango de n1 a n2, ambos inclusive
    for n in range(n1, n2 + 1):
        # Verificar si n es menor que n0 (el desplazamiento)
        if n < n0:
            # Si es menor, agregar 0 al escalón (antes del desplazamiento)
            escalon.append(0)
        else:
            # Si es mayor o igual, agregar 1 al escalón (después o en el desplazamiento)
            escalon.append(1)
    # Devolver la lista que representa la función escalón en el intervalo dado
    return escalon

# Ejemplo de uso de la función escalon_unitario
n1 = -5
n0 = 0
n2 = 5
# Llamar a la función con los parámetros especificados
resultado = escalon_unitario(n0, n1, n2)
# Imprimir el resultado
print("Función escalón unitario en el intervalo [{}, {}]: {}".format(n1, n2, resultado))
````

#PUNTO 2

````python
def funcion_rampa(n0, n1, n2, m):
    #Genera una función rampa definida en un intervalo [n1, n2] con una pendiente m.

    # Args:
     #- n0 (int): Punto de inicio del intervalo.
     #- n1 (int): Punto de inicio de la función rampa.
     #- n2 (int): Punto final del intervalo.
     #- m (float): Pendiente de la función rampa.

     #Returns:
     #- list: Lista que representa la función rampa en el intervalo [n1, n2].

    # Inicializar una lista vacía para almacenar los valores de la función rampa
    rampa = []
    # Iterar sobre cada valor en el rango de n1 a n2, ambos inclusive
    for n in range(n1, n2 + 1):
        # Calcular el valor de la función rampa en el punto actual
        # El valor es la multiplicación de la pendiente por la distancia desde n0, limitado a un mínimo de 0
        valor = max(0, m * (n - n0))
        # Agregar el valor calculado a la lista rampa
        rampa.append(valor)
    # Devolver la lista que representa la función rampa en el intervalo dado
    return rampa

# Ejemplo de uso de la función funcion_rampa
n0 = -2
n1 = 0
n2 = 5
pendiente = 2
# Llamar a la función con los parámetros especificados
resultado = funcion_rampa(n0, n1, n2, pendiente)
# Imprimir el resultado
print("Función rampa con pendiente {} en el intervalo [{}, {}]: {}".format(pendiente, n1, n2, resultado))
````

#PUNTO 3A

````python
def delta(n, n_0):
    """
    Función delta de Dirac.

    Args:
    - n (numpy array): Puntos en el eje n.
    - n_0 (int): Punto donde la función delta tiene un impulso.

    Returns:
    - numpy array: Valores de la función delta en cada punto de n.

    Utilizamos np.where para asignar 1 donde n coincide con n_0 y 0 en otros lugares

    """
    return np.where(n == n_0, 1, 0)

def x1(n):
    """
    Secuencia x1(n) = 3δ(n+1) + 5δ(n+3) + 3δ(n+2) + 3δ(n) + δ(n).

    Args:
    - n (numpy array): Vector de muestras de tiempo.

    Returns:
    - numpy array: Valores de la secuencia x1(n) para cada valor de n.

    Definimos la secuencia x1(n) utilizando la función delta definida anteriormente
    """
    return 3*delta(n, -1) + 5*delta(n, -3) + 3*delta(n, -2) + 3*delta(n, 0) + delta(n, 0)

# Vector de muestras de tiempo
n = np.arange(-10, 11)  # Creamos un vector de tiempo que va desde -10 hasta 10

# Calculamos la secuencia x1(n) utilizando la función definida
x1_n = x1(n)

# Graficamos los resultados utilizando plt.stem para una representación de la secuencia discreta
plt.stem(n, x1_n, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.title('Secuencia x1(n)')  # Agregamos un título a la gráfica
plt.xlabel('n')  # Etiqueta del eje x
plt.ylabel('x1(n)')  # Etiqueta del eje y
plt.grid(True)  # Mostrar cuadrícula
plt.show()  # Mostrar la gráfica
````

# PUNTO 3B

````python
# Definición de la función escalón unitario
def escalon_unitario(n0, n1, n2):
    """
    Genera una función escalón unitario definida en un intervalo [n1, n2] con un desplazamiento n0.

    Args:
    - n0 (int): Desplazamiento de la función escalón.
    - n1 (int): Punto de inicio del intervalo.
    - n2 (int): Punto final del intervalo.

    Returns:
    - list: Lista que representa la función escalón en el intervalo [n1, n2].
    """
    escalon = []  # Inicializar una lista vacía para almacenar los valores del escalón
    for n in range(n1, n2 + 1):  # Iterar sobre el rango de muestras de tiempo [n1, n2]
        if n < n0:  # Si n es menor que el desplazamiento n0
            escalon.append(0)  # Agregar 0 al escalón
        else:
            escalon.append(1)  # Agregar 1 al escalón
    return escalon  # Devolver la lista que representa el escalón

# Definición de la función x2(n)
def x2(n):
    """
    Genera la secuencia x2(n) = -2u(n-4) + 3u(n-2).

    Args:
    - n (list): Lista de muestras de tiempo.

    Returns:
    - list: Valores de la secuencia x2(n) para cada valor de n.
    """
    # Generar las funciones escalón u(n-4) y u(n-2)
    u_n_4 = escalon_unitario(4, min(n), max(n))  # Función escalón en n - 4
    u_n_2 = escalon_unitario(2, min(n), max(n))  # Función escalón en n - 2

    # Calcular los valores de la secuencia x2(n)
    x2_n = [-2 * u_n_4[i] + 3 * u_n_2[i] for i in range(len(n))]  # Calcula los valores de la secuencia x2(n)
    return x2_n  # Devolver los valores de la secuencia x2(n)

# Generar el rango de valores de n
n = list(range(-6, 7))

# Calcular los valores de la secuencia x2(n)
x2_n = x2(n)

# Graficar los resultados
plt.stem(n, x2_n, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.title('Secuencia x2(n)')
plt.xlabel('n')
plt.ylabel('x2(n)')
plt.grid(True)
plt.show()
````

#PUNTO 3D

````python
def u(n):
    """
    Genera la función escalón unitario u(n).

    Args:
    - n (numpy array): Vector de muestras de tiempo.

    Returns:
    - numpy array: Valores de la función escalón unitario u(n) para cada valor de n.
    """
    return np.heaviside(n, 1)

def r(n):
    """
    Genera la función rampa r(n).

    Args:
    - n (numpy array): Vector de muestras de tiempo.

    Returns:
    - numpy array: Valores de la función rampa r(n) para cada valor de n.
    """
    return np.maximum(0, n)

def x4(n):
    """
    Genera la secuencia x4(n) = 2*r(n+3) - r(n-2) - 5*u(n-3).

    Args:
    - n (numpy array): Vector de muestras de tiempo.

    Returns:
    - numpy array: Valores de la secuencia x4(n) para cada valor de n.
    """
    return 2 * r(n + 3) - r(n - 2) - 5 * u(n - 3)

# Vector de muestras de tiempo
n = np.arange(-10, 11)

# Secuencia x4(n)
x4_n = x4(n)

# Graficar el resultado
plt.stem(n, x4_n, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.title('Secuencia x4(n)')
plt.xlabel('n')
plt.ylabel('x4(n)')
plt.grid(True)
plt.show()
````

#PUNTO 4

````python
def r(n):
    """
    Genera la secuencia rampa r[n].

    Args:
    - n (numpy array): Vector de muestras de tiempo.

    Returns:
    - numpy array: Valores de la secuencia rampa r[n] para cada valor de n.
    """
    return np.maximum(0, n)

def x(n):
    """
    Genera la secuencia x[n] según las definiciones dadas.

    Args:
    - n (numpy array): Vector de muestras de tiempo.

    Returns:
    - numpy array: Valores de la secuencia x[n] para cada valor de n.
    """
    x = np.zeros_like(n)
    x[:6] = r(n[:6])
    x[6:12] = n[6:12] - 5
    x[12:] = r(n[12:] - 10)
    return x

# Vector de muestras de tiempo
n = np.arange(0, 18)

# Secuencia x(n)
x_n = x(n)

# Derivada de la secuencia x(n)
dx_n = np.gradient(x_n, 1)

# Crear subplot
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Graficar la secuencia original
axs[0].stem(n, x_n, linefmt='b-', markerfmt='bo', basefmt=' ')
axs[0].set_title('Secuencia x(n)')
axs[0].set_xlabel('n')
axs[0].set_ylabel('x(n)')
axs[0].grid(True)

# Graficar la derivada de la secuencia
axs[1].stem(n, dx_n, linefmt='r-', markerfmt='ro', basefmt=' ')
axs[1].set_title('Derivada de x(n)')
axs[1].set_xlabel('n')
axs[1].set_ylabel('dx(n)/dn')
axs[1].grid(True)

# Ajustar diseño
plt.tight_layout()
plt.show()
````

````python
# Definir la función x6(n)
def x6(n):
    # Crear un array lleno de ceros con la misma forma y tipo que el vector de muestras de tiempo n
    x6_values = np.zeros_like(n, dtype=float)

    # Iterar sobre cada elemento del vector de muestras de tiempo n
    for i, n_val in enumerate(n):
        # Calcular el término exponencial para este valor de n
        exp_term = 0.001 * np.exp(0.5 * n_val) * x(n_val)

        # Calcular el término sinusoidal para este valor de n
        sin_term = 10 * np.sin(0.05 * np.pi * n_val) * x(n_val + 2)

        # Sumar los términos exponencial y sinusoidal y asignar el resultado al valor de x6_values en esta posición
        x6_values[i] = exp_term.sum() + sin_term.sum()  # Sumar todos los elementos de los arrays

    # Devolver el array que contiene los valores de la secuencia x6(n)
    return x6_values

# Vector de muestras de tiempo
n = np.arange(-20, 21)

# Calcular la secuencia x6(n)
x6_n = x6(n)

# Graficar los resultados
plt.stem(n, x6_n, linefmt='g-', markerfmt='go', basefmt=' ')
plt.title('Secuencia x6(n)')
plt.xlabel('n')
plt.ylabel('x6(n)')
plt.grid(True)
plt.show()
````
